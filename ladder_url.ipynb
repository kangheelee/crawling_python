```python
# Define the empty list for storing the data
list_name = []
list_url = []
list_subscribe = []

# Define a function, named "ladder_url"
## 1. Call the Chrome driver in local path
## 2. Time delay for safty programming (5 seconds)
## 3. Get Web-site Url where you want to crawling
## 4. Repeat the automatic actions until range in your epochs 
def ladder_url(url):
    # Call Chrome Drive
    driver= webdriver.Chrome('/Users/Jaehyung/Desktop/chromedriver_win32/chromedriver')
    
    # Wait for 5 seconds
    driver.implicitly_wait(5)
    
    # Get Web-site URL
    driver.get(url)
    
    # For page down action (option)
    count = 1
    
    # 
    for i in range(1,61):
        
        driver.find_element_by_xpath('//*[@id="top"]/div[3]/div[1]/div/div[1]/div/div[10]').click()
        driver.find_element_by_xpath('//*[@id="top"]/div[3]/div[2]/div/div[2]').click()
        driver.find_element_by_xpath('//*[@id="top"]/div[3]/div[2]/div/div[3]').click()
        sub_count = driver.find_element_by_xpath('//*[@id="top"]/div[3]/div[3]/div/div[2]/div/div/div[2]/div')
        move= ActionChains(driver)
        move.click_and_hold(sub_count).move_by_offset(10,0).release().perform()
        driver.find_element_by_xpath('//*[@id="top"]/div[3]/div[3]/div/button').click()
        driver.find_element_by_xpath('//*[@id="top"]/div[3]/div[2]/div/div[4]/div').click()
        driver.find_element_by_xpath('/html/body/div[3]/div[1]/div[1]/ul/li[1]').click()
        
    
        body = driver.find_element_by_tag_name('body')
        
        
        if i % 20 == 0:
            count = count + 1
            
        
        for idx in range(count):
            body.send_keys(Keys.PAGE_DOWN)
            time.sleep(1)
            body.send_keys(Keys.PAGE_DOWN)
            time.sleep(1)
        
        
        
       
       
        
        
        
        if i < 4:
            driver.find_element_by_xpath('//*[@id="top"]/div[3]/div[4]/div/div[1]/div[%d]' %i ).click()
            youtuber_name = driver.find_element_by_xpath('//*[@id="top"]/div[3]/div[2]/div[1]/div[2]/div[1]').text
            subscribe = driver.find_element_by_xpath('//*[@id="top"]/div[3]/div[2]/div[1]/div[2]/div[2]/div[2]/span').text
            list_subscribe.append(subscribe)
            list_name.append(youtuber_name)
            list_url.append(driver.current_url)
            #driver.find_element_by_xpath('//*[@id="top"]/div[3]/div[2]/div[2]/div').click()
            
            #driver.window_handles
            #last_tab = driver.window_handles[-1]
            #driver.switch_to.window(window_name = last_tab)
            
            #driver.close()
          
            #first_tab = driver.window_handles[0]
            #driver.switch_to.window(window_name = first_tab)
            driver.back()
        
        elif i >= 4: 
            driver.find_element_by_xpath('//*[@id="top"]/div[3]/div[4]/div/div[2]/div[%d-2]' %i).click()
            youtuber_name = driver.find_element_by_xpath('//*[@id="top"]/div[3]/div[2]/div[1]/div[2]/div[1]').text
            subscribe = driver.find_element_by_xpath('//*[@id="top"]/div[3]/div[2]/div[1]/div[2]/div[2]/div[2]/span').text
            list_subscribe.append(subscribe)
            list_name.append(youtuber_name)
            list_url.append(driver.current_url)
            #driver.find_element_by_xpath('//*[@id="top"]/div[3]/div[2]/div[2]/div').click()
            
        
            #driver.window_handles
            #last_tab = driver.window_handles[-1]
            #driver.switch_to.window(window_name = last_tab)
            #list_url.append(driver.current_url)
            #driver.close()
            
            #first_tab = driver.window_handles[0]
            #driver.switch_to.window(window_name = first_tab)
            driver.back()
            
        
        



        
```
